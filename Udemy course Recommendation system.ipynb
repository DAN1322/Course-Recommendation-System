```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import neattext.functions as nfx

import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
df = pd.read_csv("udemydata.csv")
df.head()
```

```python
list(dir(nfx))
```

```python
list((df['course_title']))
```

```python
df['cleantitle'] = df['course_title'].apply(nfx.remove_stopwords)
df['cleantitle'] = df['cleantitle'].apply(nfx.remove_special_characters)
list(df['cleantitle'])
```

```python
# count vectorizer is a tool used in NLP to convert a collection of text documents into a matrix of token (word) counts. it is 
#commonly used when preparig text data for ML models
#in simple , it turns words into numbers which is what models need to work with
countvectorizer = CountVectorizer()
CV = countvectorizer.fit_transform(df['cleantitle'])
CV
```

```python
df_cv_words = pd.DataFrame(CV.todense(), columns=countvectorizer.get_feature_names())
df_cv_words
```

```python
# cosine similarity measures how similar 2 vectors are - based on the angle between them - not their magnitude
# in NLp, it's commonly used to compare text documents after converting them into vectors(eg using countvectorizer or TFIDF vectorizer)
cosine_similarity = cosine_similarity(CV)
cosine_similarity
```

```python
courseindex = pd.Series(df.index,index = df['course_title'])
courseindex.drop_duplicates()
courseindex
```

```python
temp = df[df['course_title'].str.contains('Trading')]
temp.head()
```

```python
top10 = temp.sort_values(by = 'num_subscribers', ascending=False).head(10)
top10
```

```python
index = courseindex['How To Maximize Your Profits Trading Options']
```

```python
scores = list(enumerate(cosine_similarity[index]))
scores
```

```python
sortedscore = sorted(scores,key = lambda x:x[1],reverse=True)
sortedscore
```

```python
sorted_indices = [i[0] for i in sortedscore[1:]]
sorted_values = [i[1] for i in sortedscore[1:]]
sorted_values
```

```python
result = df.iloc[sorted_indices]
result
```

```python
result['similaritscore'] = np.array(sorted_values)
result
```

```python
final_df = result[['cleantitle','similaritscore']]
final_df
          
```

```python
def recommend(title, n=10):
    courseindex = pd.Series(df.index, index=df['course_title']).drop_duplicates()
    index = courseindex[title]
    scores = list(enumerate(cosine_similarity[index]))
    sortedscore = sorted(scores,key = lambda x:x[1],reverse=True)
    sorted_indices = [i[0] for i in sortedscore[1:]]
    sorted_values = [i[1] for i in sortedscore[1:]]
    result = df.iloc[sorted_indices]
    result['similarityscore'] = np.array(sorted_values)
    final_recommended_courses = result[['course_title', 'similarityscore', 'url', 'price', 'num_subscribers']]
    return final_recommended_courses.head(n)

```

```python
recommend('Forex Trading A-Z™ - With LIVE Examples of Forex Trading', 10)
```
